## QA

```python
message = "get information for ProjectEnigmaHackathon/hack-001 repository"
orchestrator = get_orchestrator()
```

```python
classification = await orchestrator.classify_workflow(message)

workflow_type = classification["workflow_type"]
            
print(f"Orchestrator classified message as: {workflow_type} (confidence: {classification.get('confidence', 'N/A')})")
print(f"Reasoning: {classification.get('reasoning', 'N/A')}")
```

```python
workflow_manager = get_workflow_manager_by_type(workflow_type)
```

```python
from app.models.api import (
    ChatRequest)
request = ChatRequest(message=message,
    repositories=[],
    release_type= "release",
    sprint_name="",
    fix_version="",
    session_id="")
```

```python
from app.api.endpoints.chat import create_initial_workflow_state
initial_state = create_initial_workflow_state(request, workflow_type)
initial_state
```

```python
workflow_id = await workflow_manager.start_workflow(initial_state)
workflow_id
```

```python
status = workflow_manager.get_workflow_status(workflow_id)
if status:
    print(f"\nWorkflow status:")
    print(f"  workflow_id: {status.get('workflow_id', 'Unknown')}")
    print(f"  metadata: {status.get('metadata', 'Unknown')}")
    print(f"  state: {status.get('state', 'Unknown')}")
    print(f"  is_running: {status.get('is_running', 'Unknown')}")
```

```python
import asyncio

async def print_workflow_stream():
    status_stream = workflow_manager.get_workflow_stream(workflow_id)
    async for x in status_stream:
        print(x)

await print_workflow_stream()
```
